```python
import pandas as pd
import matplotlib.pyplot as plt

#Loading the pizza data
file_path = r"C:\Users\HUBERT\Desktop\Datasets\pizza.csv"
df = pd.read_csv(file_path)

#first few rows of the data
print(df.head())

#summary statistics
print(df.describe())

#Check for missing values
print(df.isnull().sum())

# Handling missing values by filling them with the mean for numeric columns 
# using numeric_only=True to avoid errors with non-numeric columns.
df.fillna(df.mean(numeric_only=True), inplace=True)

# Alternatively, you could remove rows with missing values
# df.dropna(inplace=True)

# Display the first few rows to verify changes
print(df.head())

# Example analysis: distribution of pizza sizes
pizza_sizes = df['size'].value_counts()
print(pizza_sizes)

# Example analysis: average price by size
average_price_by_size = df.groupby('size')['price'].mean()
print(average_price_by_size)

# Plot distribution of pizza sizes
plt.figure(figsize=(10, 6))
pizza_sizes.plot(kind='bar', color='skyblue')
plt.xlabel('Size')
plt.ylabel('Count')
plt.title('Distribution of Pizza Sizes')
plt.show()

# Plot average price by size
plt.figure(figsize=(10, 6))
average_price_by_size.plot(kind='bar', color='lightgreen')
plt.xlabel('Size')
plt.ylabel('Average Price')
plt.title('Average Price by Pizza Size')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = r"C:\Users\HUBERT\Desktop\Datasets\pizza.csv"
df = pd.read_csv(file_path)

# Display the first few rows of the data
print(df.head())

# Display summary statistics
print(df.describe())

# Check for missing values
print(df.isnull().sum())

# Handle missing values by filling with mean for numeric columns
df.fillna(df.mean(numeric_only=True), inplace=True)

# Alternatively, you could remove rows with missing values
# df.dropna(inplace=True)

# Display the first few rows to verify changes
print(df.head())

# Example analysis: distribution of pizza sizes
pizza_sizes = df['size'].value_counts()
print(pizza_sizes)

# Example analysis: average price by size
average_price_by_size = df.groupby('size')['price'].mean()
print(average_price_by_size)

# Select only numeric columns for correlation
numeric_cols = df.select_dtypes(include=['number']).columns
correlation_matrix = df[numeric_cols].corr()
print(correlation_matrix)

# Plot distribution of pizza sizes
plt.figure(figsize=(10, 6))
pizza_sizes.plot(kind='bar', color='skyblue')
plt.xlabel('Size')
plt.ylabel('Count')
plt.title('Distribution of Pizza Sizes')
plt.show()

# Plot average price by size
plt.figure(figsize=(10, 6))
average_price_by_size.plot(kind='bar', color='lightgreen')
plt.xlabel('Size')
plt.ylabel('Average Price')
plt.title('Average Price by Pizza Size')
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Box plot for price by size
plt.figure(figsize=(10, 6))
sns.boxplot(x='size', y='price', data=df)
plt.xlabel('Size')
plt.ylabel('Price')
plt.title('Box Plot of Price by Size')
plt.show()
```
