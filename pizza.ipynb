```python
import pandas as pd
import matplotlib.pyplot as plt

#Loading the pizza data
file_path = r"C:\Users\HUBERT\Desktop\Datasets\pizza.csv"
df = pd.read_csv(file_path)

#first few rows of the data
print(df.head())

#summary statistics
print(df.describe())

#Check for missing values
print(df.isnull().sum())

# Handle missing values by filling with mean for numeric columns
df.fillna(df.mean(numeric_only=True), inplace=True)

# Alternatively, you could remove rows with missing values
# df.dropna(inplace=True)

# Display the first few rows to verify changes
print(df.head())

# Example analysis: distribution of pizza sizes
pizza_sizes = df['size'].value_counts()
print(pizza_sizes)

# Example analysis: average price by size
average_price_by_size = df.groupby('size')['price'].mean()
print(average_price_by_size)

# Plot distribution of pizza sizes
plt.figure(figsize=(10, 6))
pizza_sizes.plot(kind='bar', color='skyblue')
plt.xlabel('Size')
plt.ylabel('Count')
plt.title('Distribution of Pizza Sizes')
plt.show()

# Plot average price by size
plt.figure(figsize=(10, 6))
average_price_by_size.plot(kind='bar', color='lightgreen')
plt.xlabel('Size')
plt.ylabel('Average Price')
plt.title('Average Price by Pizza Size')
plt.show()
```
