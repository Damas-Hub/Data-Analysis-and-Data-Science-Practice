Visualizing the dataset


```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
 file_path = r"C:\Users\HUBERT\Desktop\Datasets\Weather.csv"
 df = pd.read_csv( file_path)
 print(df)
```

```python
# Plot temperature over time
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Temp_C'], label='Temperature (C)')
plt.xlabel('Date/Time')
plt.ylabel('Temperature (C)')
plt.title('Temperature Trends Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot humidity over time
plt.figure(figsize=(14, 7))
plt.plot(df.index, df['Rel Hum_%'], label='Relative Humidity (%)', color='blue')
plt.xlabel('Date/Time')
plt.ylabel('Relative Humidity (%)')
plt.title('Humidity Trends Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot distribution of wind speed
plt.figure(figsize=(20, 10))
plt.hist(df['Wind Speed_km/h'], bins=30, edgecolor='black', color='green')
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Frequency')
plt.title('Wind Speed Distribution')
plt.grid(True)
plt.show()
```

```python
# Scatter plot of temperature vs. humidity
plt.figure(figsize=(14, 7))
plt.scatter(df['Temp_C'], df['Rel Hum_%'], alpha=0.5, color='red')
plt.xlabel('Temperature (C)')
plt.ylabel('Relative Humidity (%)')
plt.title('Temperature vs. Humidity')
plt.grid(True)
plt.show()
```

```python
# Box plot for temperature
plt.figure(figsize=(14, 7))
plt.boxplot(df['Temp_C'], vert=False)
plt.xlabel('Temperature (C)')
plt.title('Box Plot for Temperature')
plt.grid(True)
plt.show()
```

```python
# Histogram for visibility
plt.figure(figsize=(14, 7))
plt.hist(df['Visibility_km'], bins=30, edgecolor='black', color='purple')
plt.xlabel('Visibility (km)')
plt.ylabel('Frequency')
plt.title('Visibility Distribution')
plt.grid(True)
plt.show()
```
